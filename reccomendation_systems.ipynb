# Popularity based Recommender Systems

```python
ratings_with_books_df = ratings_df.merge(books_df,on='ISBN')
ratings_with_books_df.shape
```

```python
ratings_with_books_df.head(5)
```

```python
num_ratings_df = ratings_with_books_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_ratings_df.head(5)
```

```python
avg_ratings_df = ratings_with_books_df.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_ratings_df.head(5)
```

```python
popularity_df = num_ratings_df.merge(avg_ratings_df,on='Book-Title')
popularity_df = popularity_df[popularity_df['num_ratings'] > 250]
popularity_df = popularity_df.sort_values('avg_ratings',ascending=False)
# we need top 50 books
popularity_df = popularity_df.head(50) 
```

```python
popularity_books_df = popularity_df.merge(books_df,on='Book-Title')
popularity_books_df.drop_duplicates("Book-Title",inplace=True)
popularity_books_df.head(5)
```

```python
popularity_books_df = popularity_books_df[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
popularity_books_df = popularity_books_df.reset_index()
popularity_books_df = popularity_books_df.drop('index',axis=1)
popularity_books_df
```

# Collaborative Filtering based Recommendation Systems

```python
# fetch user who have given rating more than 200
x = ratings_with_books_df.groupby('User-ID').count()['Book-Rating'] > 200
user_rated_200 = x[x].index  
```

```python
filtered_ratings = ratings_with_books_df[ratings_with_books_df['User-ID'].isin(user_rated_200)]
ratings_with_books_df.shape
```

```python
# Fetch the books who count of rating is greater than 50
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] > 50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings.shape
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
# function to recommend top 5 books
import numpy as np
def recommend_books(book_name):
    suggestions = [ ]
    #fetch index from books
    index = np.where(pt.index==book_name)[0][0]
    # get the top 5 similar items based on
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        suggestions.append(pt.index[i[0]])
    return suggestions
```

```python
recommend_books('Harry Potter and the Chamber of Secrets (Book 2)')
```
